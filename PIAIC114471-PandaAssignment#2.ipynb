```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
us_census = pd.concat([pd.read_csv(f, index_col=0) for f in glob.glob('states*.csv')], ignore_index=True)
```

```python
us_census.Income = us_census.Income.str.lstrip('$').astype('float')
```

```python
us_census.head()
```

```python
us_census[['Men', 'Women']] = us_census.GenderPop.str.split('_', expand=True)
```

```python
us_census.drop(['GenderPop'], axis=1, inplace=True)
us_census.head()
```

```python
us_census.Men = us_census.Men.str.rstrip('M')
us_census.Women = us_census.Women.str.rstrip('F')
us_census.head()
```

```python
us_census.Women = pd.to_numeric(us_census.Women)
us_census.Men = pd.to_numeric(us_census.Men)
```

```python
plt.scatter(us_census.Women, us_census.Income)
plt.show()
```

```python
us_census.Women.fillna(value=us_census.TotalPop-us_census.Men, inplace=True)
```

```python
plt.scatter(us_census.Women, us_census.Income)
plt.show()
```

```python
us_census.duplicated()
```

```python
us_census.drop_duplicates(inplace=True)
```

```python
plt.scatter(us_census.Women, us_census.Income, color='red')
plt.title('Scatter plot')
plt.show()
```

```python
display(us_census.head())
print(us_census.columns)
us_census.info()
```

```python
def setting_races(x):
    x = x.str.rstrip('%').astype('float64')
    x = round((us_census.TotalPop * x)/100, 2)
    return x
```

```python
us_census.loc[: ,'Hispanic':'Pacific'] = us_census.loc[: ,'Hispanic' : 'Pacific'].apply(setting_races)
```

```python
us_census.head(10)
```

```python
us_census.Pacific.fillna(value=us_census.Pacific.sum()/us_census.Pacific.size, inplace=True)
```
